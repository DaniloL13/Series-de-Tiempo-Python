```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.arima_model import ARMA
from scipy.stats.distributions import chi2 
import seaborn as sns
sns.set()
```

### DATA

```python
data = pd.read_csv("/content/Indice.csv") 
df = data.copy()
df.date = pd.to_datetime(df.date, dayfirst = True)
df.set_index("date", inplace=True)
df=df.asfreq('b')
df=df.fillna(method='ffill')
```

```python
df['market_value']=df.ftse
```

```python
del df['spx']
del df['dax']
del df['ftse']
del df['nikkei']
size = int(len(df)*0.8)
df, df_test = df.iloc[:size], df.iloc[size:]
```

### RETORNOS

```python
df['returns'] = df.market_value.pct_change(1)*100
```

### MODELOS ARMA(p,q)

#### Test LLR 

```python
def LLR_test(mod_1, mod_2, DF = 1):
    L1 = mod_1.llf
    L2 = mod_2.llf
    LR = (2*(L2-L1))    
    p = chi2.sf(LR, DF).round(3)
    return p
```

```python
import warnings
warnings.filterwarnings("ignore")
```

#### ARMA(1,1)

```python
model_ar_1_ma_1 = ARMA(df.returns[1:], order=(1,1))
ajuste_ar_1_ma_1 = model_ar_1_ma_1.fit()
ajuste_ar_1_ma_1.summary()
```

```python
model_ar_1 = ARMA(df.returns[1:], order = (1,0))
model_ma_1 = ARMA(df.returns[1:], order = (0,1))
ajuste_ar_1 = model_ar_1.fit()
ajuste_ma_1 = model_ma_1.fit()
```

```python
print("\nARMA vs AR ", LLR_test(ajuste_ar_1, ajuste_ar_1_ma_1))
print("\nARMA vs MA ", LLR_test(ajuste_ma_1, ajuste_ar_1_ma_1))
```
